```python
data1 = {'Name':['Jai', 'Anuj', 'Jai', 'Princi', 
                 'Gaurav', 'Anuj', 'Princi', 'Abhi'], 
        'Age':[27, 24, 22, 32, 
               33, 36, 27, 32], 
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj',
                   'Jaunpur', 'Kanpur', 'Allahabad', 'Aligarh'], 
        'Qualification':['Msc', 'MA', 'MCA', 'Phd',
                         'B.Tech', 'B.com', 'Msc', 'MA']} 
myData = pd.DataFrame(data1)
myData
```

```python
## Grouping by name
myData.groupby('Name')
print(myData.groupby('Name').groups)
```

```python
## printing the first entries in each group
gk  = myData.groupby('Name')
gk.first()
```

```python
## grouping by multiple keys
groups = myData.groupby(['Name','Qualification'])
print(groups.groups)
```

```python
## grouping by sorting keys
data2 = {'Name':['Jai', 'Anuj', 'Jai', 'Princi', 
                 'Gaurav', 'Anuj', 'Princi', 'Abhi'], 
        'Age':[27, 24, 22, 32, 
               33, 36, 27, 32], } 
df2 = pd.DataFrame(data2)
#print(df2.groupby('Name').sum())
#when we sort
print(df2.groupby(['Name'],sort = True).sum()) # sort can take true or false, use false for speedup
```

## Pandas Merging, Joining, and Concatenating

```python
### Concatenating DataFrame
```

```python
### using concat
data3 = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32], 
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']}
data4 = {'Name':['Abhi', 'Ayushi', 'Dhiraj', 'Hitesh'], 
        'Age':[17, 14, 12, 52], 
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
df3 = pd.DataFrame(data3, index=[0,1,2,3])
df4 = pd.DataFrame(data4, index=[4,5,6,7])
#print(df3, "\n\n", df4)
frames = [df3,df4]
mergeDf1 = pd.concat(frames)
mergeDf1
```

```python
#Concatenating DataFrame by setting logic on axes :
mergeDf2 = pd.concat([df3,df4], axis=1, join='inner')
mergeDf2
mergeDf3 = pd.concat([df3,df4], axis=1, join = 'outer')
mergeDf3
mergeDf4 = pd.concat([df3,df4],axis =1)
mergeDf4 = mergeDf4.reindex(df4.index)
```

```python
# Concatenating DataFrame using .append()
res = df3.append(df4)
res
```

```python
#Concatenating DataFrame by ignoring indexes :
res1 = pd.concat(frames,ignore_index = True)
res1
```

```python
# Concatenating DataFrame with group keys :
res2 = pd.concat(frames, keys=['x','y'])
res2
```

```python
# Concatenating with mixed ndims :
s1 = pd.Series([1000,2000,2000,4000], name = 'Salary')
res3 = pd.concat([df3,s1],axis=1)
res3
```

## Working with date and time using Pandas

```python
# creating a date dataframe with frequency
dateDf = pd.date_range('1/1/2011',periods = 10, freq = 'H')
now = datetime.datetime.now()
now.month, now.year
```

```python
# Break data and time into separate features
timeDf = pd.DataFrame()
timeDf['date'] = dateDf
timeDf['year'] = timeDf['date'].dt.year
timeDf['month'] = timeDf['date'].dt.month
timeDf['day'] = timeDf['date'].dt.day
timeDf['minute'] = timeDf['date'].dt.minute
timeDf[:5]
```
