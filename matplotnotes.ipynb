# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#simple line graphs
x = np.array([0,4,5,6,7,8,9])
y = np.array([0,20,30,40,50,60,70])
plt.plot(x,y)
plt.show()
```

```python
#plotting without lines--just points
x1 = np.array([1,8])
y1 = np.array([3,10])
plt.plot(x1,y1,'o')
```

```python
#plotting with makers
y2 = np.array([3,8,10,15])
plt.plot(y2,marker='*')
```

```python
#using the shothand prop --with marker size(ms),marker color(mec)
plt.plot(y2, 'o:m', ms = 10) #marker|line|color
plt.show()
```

```python
#maker face,edge color
plt.plot(y2,marker='o',ms = 15, mec = 'r', mfc = 'b') #color code can be hexadecimal
```

```python
#linestyle(ls), linewidth
plt.plot(y2, ls = '--',marker = '*',c='m',linewidth = 5)
```

```python
#multiple line --assuming default x values
y3 = np.array([6,2,7,11])
plt.plot(y2)
plt.plot(y3)
```

```python
#creating title, location, specifying grids
plt.plot(y2)
plt.title('Sports data', loc = 'left')
plt.xlabel('Average time')
plt.ylabel('Number of players')
plt.grid(color = 'green',axis = 'x', linestyle ='--', linewidth = 0.5) 
```

## Subplots --drwaing multiple plots in the same figure


```python
#draw two plots side by side
x4 = np.array([0, 1, 2, 3])
y4 = np.array([3, 8, 1, 10])
plt.subplot(1,2,1) #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x4,y4)

x5 = np.array([0, 1, 2, 3])
y5 = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2) #the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(x5,y5)
```

```python
#2 plots on top of each other
plt.subplot(2,1,1)
plt.plot(x4,y4)

plt.subplot(2,1,2)
plt.plot(x5,y5)
```

```python
#drawing 4 plots
plt.subplot(2,2,1)
plt.plot(x4,y4)
plt.title('Sales')

plt.subplot(2,2,2)
plt.plot(x5,y5)
plt.title('Income')

plt.subplot(2,2,3)
plt.plot(y2)
plt.title('Profits')

plt.subplot(2,2,4)
plt.plot(x,y)
plt.title('Losses')

plt.suptitle('My Business')
```

```python
### Scatter plots
a = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
b = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(a, b)

a1 = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
b1 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(a1, b1, color = 'm')

#The X-axis shows how old the car is.
#The Y-axis shows the speed of the car when it passes.
```

```python
#colors and transparency
size = np.array([20,50,100,200,500,800,60,90,10,300,600,800,75,100,20])
plt.scatter(a1,b1, s = size, alpha=0.5)
```

```python
# alt
a3 = np.random.randint(100, size=(100))
b3 = np.random.randint(100, size= (100))

colors = np.random.randint(100, size=(100))
sizes = np.random.randint(100, size=(100))

plt.scatter(a3,b3, c = colors, alpha = 0.5, cmap = 'nipy_spectral', s = sizes )
plt.colorbar()
```

```python
myArr = np.array([20,15,10,5,40,10])
myLabels = ['apples','pines','mangoes','bananas','peach','dates']
plt.pie(myArr, labels = myLabels)
```

```python
#exploding parts of the pie
myExplode = [0.2,0,0,0,0.4,0]
plt.pie(myArr, labels = myLabels, explode = myExplode, shadow =True)
plt.legend(title = 'Fruits') #adds labels at end
```
